# SEARCHING DATA OSN 2024 
search mengunakan algoritma linier search data dibedakan menjadai 2 klasifikasi :

1. Data terurut (linier search + shell sort)
2. Data belum terurut (hanya linier search)

# Data terurut (linier search + shell sort)

```python
import pandas as pd
import time
from IPython.display import display

df = pd.read_csv("osn.csv")

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# =========================================================
# SHELL SORT
# =========================================================

def shell_sort(data, kolom):

    n = len(data)
    gap = n // 2

    while gap > 0:

        for i in range(gap, n):

            temp = data[i]
            j = i

            while (
                j >= gap and
                str(data[j-gap][kolom]).lower() >
                str(temp[kolom]).lower()
            ):

                data[j] = data[j-gap]
                j -= gap

            data[j] = temp

        gap //= 2

    return data

# =========================================================
# LINEAR SEARCH
# =========================================================

def linear_search(data, kolom, keyword):

    hasil = []

    keyword = keyword.lower()

    for item in data:

        if str(item[kolom]).lower() == keyword:
            hasil.append(item)

    return pd.DataFrame(hasil)

# =========================================================
# PROGRAM UTAMA
# =========================================================

print("=" * 80)
print("PROGRAM SHELL SORT + LINEAR SEARCH")
print("=" * 80)

print("Jumlah Data :", len(df))

print("\n1. Cari Provinsi")
print("2. Cari Kab/Kota")
print("3. Cari Bidang")
print("4. Cari Medali")

pilihan = input("\nMasukkan pilihan : ")

if pilihan == "1":

    kolom = "Provinsi"
    keyword = input("Masukkan nama provinsi : ")

elif pilihan == "2":

    kolom = "Kab/Kota"
    keyword = input("Masukkan nama kota/kabupaten : ")

elif pilihan == "3":

    kolom = "Bidang"
    keyword = input("Masukkan bidang lomba : ")

elif pilihan == "4":

    kolom = "Medali"
    keyword = input("Masukkan medali : ")

else:

    print("Pilihan tidak valid.")
    exit()

# =========================================================
# SORTING + SEARCHING
# =========================================================

data = df.to_dict("records")

start = time.time()

data = shell_sort(data, kolom)
hasil = linear_search(data, kolom, keyword)

end = time.time()

# =========================================================
# OUTPUT
# =========================================================

if len(hasil) > 0:

    hasil.index = range(1, len(hasil) + 1)

    print("\nDATA DITEMUKAN")
    print("=" * 120)

    display(hasil)

    print("=" * 120)
    print("Jumlah Data :", len(hasil))

else:

    print("\nData tidak ditemukan.")

# =========================================================
# WAKTU EKSEKUSI
# =========================================================

print("\nSTATISTIK PROGRAM")
print("=" * 80)
print(f"Waktu Eksekusi Total : {end - start:.6f} detik")
```

# Data belum terurut (hanya linier search)

```python
import pandas as pd
import time
from IPython.display import display

df = pd.read_csv("osn.csv")

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

def tampilkan_hasil(dataframe):

    if len(dataframe) > 0:

        dataframe = dataframe.copy()
        dataframe.index = range(1, len(dataframe) + 1)

        print("\nDATA DITEMUKAN")
        print("=" * 120)

        display(dataframe)

        print("=" * 120)
        print("Jumlah Data Ditemukan :", len(dataframe))

    else:
        print("\nData tidak ditemukan.")

def linear_search(df, kolom, keyword):

    hasil = []

    for i in range(len(df)):

        nilai = str(df.iloc[i][kolom])

        if keyword.lower() in nilai.lower():
            hasil.append(df.iloc[i])

    return pd.DataFrame(hasil)

print("=" * 100)
print("PROGRAM LINEAR SEARCH DATA OSN")
print("=" * 100)

print("Jumlah Data :", len(df))

# =========================================================
# MENU
# =========================================================

print("\nPILIH JENIS PENCARIAN")
print("1. Kab/Kota")
print("2. Provinsi")
print("3. Bidang Lomba")
print("4. Medali")

pilihan = input("\nMasukkan pilihan (1-4): ")

# =========================================================
# PILIH KOLOM BERDASARKAN MENU
# =========================================================

if pilihan == "1":

    kolom = "Kab/Kota"
    keyword = input("\nMasukkan nama Kab/Kota : ")

elif pilihan == "2":

    kolom = "Provinsi"
    keyword = input("\nMasukkan nama Provinsi : ")

elif pilihan == "3":

    kolom = "Bidang Lomba"
    keyword = input("\nMasukkan nama Bidang Lomba : ")

elif pilihan == "4":

    kolom = "Medali"
    keyword = input("\nMasukkan jenis Medali : ")

else:

    print("Pilihan tidak valid.")
    exit()

# =========================================================
# PROSES PENCARIAN
# =========================================================

start = time.time()

hasil = linear_search(df, kolom, keyword)

end = time.time()

# =========================================================
# OUTPUT
# =========================================================

tampilkan_hasil(hasil)

print("\nWaktu Eksekusi :", round(end - start, 6), "detik")
```
